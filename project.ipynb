```python
import pandas as pd
from sqlalchemy import create_engine

# Read the CSV file
df = pd.read_csv(r'C:\Users\Anant Jaya Yadav\Desktop\anant jupyter notebook\orders.csv')

# Convert the order_date column to datetime format
df['order_date'] = pd.to_datetime(df['order_date'], format='%Y-%m-%d')

# Create a connection to the PostgreSQL database
engine = create_engine('postgresql://postgres:password@localhost:5432/projectpython')

# Write the DataFrame to PostgreSQL
df.to_sql('df_orders', engine, if_exists='replace', index=False)
```
